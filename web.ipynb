Extracting html code of any website

```python
from bs4 import BeautifulSoup
import requests
```

```python
a='https://timesofindia.indiatimes.com/'
page=requests.get(a)
```

```python
code=BeautifulSoup(page.text,'html')
print(code)
```

```python
t=code.find('title')
print(t)
```

```python
print(t.text)
```

```python
b='https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1'
pg=requests.get(b)
```

```python
cd=BeautifulSoup(pg.text,'html')
print(cd)
```

```python
t1=cd.find('table')
print(t1)
```

```python
th=t1.find_all('th')
print(th)
head=[]
for i in th:
  head.append(i.text)
print(head)
#list comprehension
#syntax:list_name=[o/p forloop]
head=[i.text for i in th]
print(head)
```

```python
#creating dataframe
import pandas as pd
df=pd.DataFrame(columns=head)
print(df)
```

```python
rows=t1.find_all('tr')
#print(tr)
ind=0
for i in rows[1:6]:
  erd=i.find_all('td')
  rd=[j.text for j in erd]
  print(rd)
  df.loc[ind]=rd
  ind=ind+1
print(df)
```
